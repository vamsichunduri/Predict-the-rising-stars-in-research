```python
#Simpson's Diversity Index
total=0
for key, value in NO_OF_PUBLICATIONS.items():
     total+=value;
total
 

div=[]

for key,value in NO_OF_PUBLICATIONS.items():
        D=0
        D=1-(value/total)
        div.append(D)
```

```python
NO_PUB['diversity']=div
NO_PUB
```

```python
author_dict = {}
for names in NO_OF_PUBLICATIONS.keys():
    r=raw_data[raw_data['authors']== names]['journal']
    author_dict[names]=r       
               
               
```

```python
def common_journals(x,y):
    ctr=0
    for i in author_dict[x]:
        for j in author_dict[y]:
            if i==j:
                ctr+=1
                
    return ctr            
    
    
```

```python
common_journals('N. Gupta','L. J. Frewer')
```

```python
total_list=[]
for aths in NO_OF_PUBLICATIONS.keys():
    commn_list=[]
    for athr in NO_OF_PUBLICATIONS.keys():
        VAL=common_journals(aths,athr)
        if VAL>1:
            commn_list.append(VAL)
            
    total_list.append(commn_list)        
```

```python
for nums in commn_list:
    if nums>1:
        print (nums)
```

```python
array_list=[]
import numpy as np
for values in total_list:
    arr=np.array(values)
    arr_mean=arr.mean()
    array_list.append(arr_mean)
    
```
